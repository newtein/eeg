```python
import os
import pickle
```

```python
path="alcoholDbFull/"

newpathA="processedData/alcohol/"
newpathC="processedData/control/"
```

```python
trails=os.listdir(path)

memory={}
for trail in trails:
    row=trail.split("_")
    print(row)
    patient=row[0]
    stimulus=row[1]+row[2]
    if patient not in memory:
        memory={}
        memory[patient]=[]
    if stimulus not in memory[patient]:
        memory[patient].append(stimulus)
        temp=pickle.load(open(path+trail,"rb"))
        if patient[3]=='a':
            pickle.dump(temp,open(newpathA+patient+"_"+stimulus,"wb"))
        else:
            pickle.dump(temp,open(newpathC+patient+"_"+stimulus,"wb"))
        
        
        
        
        
    
    
```
