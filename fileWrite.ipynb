```python
import pickle
```

```python
def Upto2(s):
    return "{0:.2f}".format(s)
```

```python
bwaves=["Wave","Gamma","Beta","Alpha","Theta","Delta"]
#bwaves=["Wave","Peel"]
channel=['F','C','P','O','T']
lobes=['Frontal','Central', 'Parietal','Occipital','Temporal']
f=open("waveAndPeel/ControlactivationValues.csv","w")
for bwave in bwaves:
    path="activation/brainWaves/"
    if bwave=="Wave":
        activationC=pickle.load(open(path+"controlActivation.pickle","rb"))
        activationStdC=pickle.load(open(path+"controlActivationStd.pickle","rb"))
    else:
    
        activationC=pickle.load(open(path+"control_{}Activation.pickle".format(bwave),"rb"))
        activationStdC=pickle.load(open(path+"control_{}ActivationStd.pickle".format(bwave),"rb"))
    
    
    for stimulus in activationC:
        f.write(bwave+": Percent Activation : Control-{}\n".format(stimulus))
        f.write(",Frontal,Central,  Parietal,Occipital,Temporal\n")
        l=['-','-','-','-','-','-']  
        ncount=0
        for i in range(len(channel)):
            count=0
            l[0]=lobes[ncount]
            
            count=1
            for j in range(0,i+1):
                try:
                    key=channel[j]+'-'+channel[i]
                    l[count]=Upto2(activationC[stimulus][key]*100)+"+-"+Upto2(activationStdC[stimulus][key]*100)
                    count+=1
                except:
                    key=channel[i]+'-'+channel[j]
                    l[count]=Upto2(activationC[stimulus][key]*100)+"+-"+Upto2(activationStdC[stimulus][key]*100)
                    count+=1
            ncount+=1
                    
            print("-->")
            f.write(",".join([str(i) for i in l])+"\n")
        #f.write("\n")
    f.write("\n")
f.close()
```
